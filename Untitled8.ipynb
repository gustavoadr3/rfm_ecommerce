```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_rfm = pd.read_csv('/content/data.csv', encoding='latin1')
```

```python
df_rfm.describe()
```

```python
# Conhecendo os dados

df_rfm.head()

df_rfm.info()
```

```python
# Valores Nulos

valores_nulos = df_rfm.isnull()

# Somar os Nulos

soma_nulos = valores_nulos.sum()

print(soma_nulos)
```

```python
# Removendo os nulos

df_rfm = df_rfm.dropna()

# Conferindo se foram removidos certos

valores_nulos = df_rfm.isnull()

soma_nulos = valores_nulos.sum()

print(soma_nulos)
```

```python
df_rfm.info()
```

```python
# Buscando todos os produtos com preço <= 0

df_rfm[df_rfm['UnitPrice'] <= 0]

```

```python
# Buscando todos os produtos com quantidade <= 0

df_rfm[df_rfm['Quantity'] <= 0]
```

```python
# Filtrando o DF p/ conter somente os dados acima de 0

df_rfm = df_rfm[(df_rfm['UnitPrice'] > 0) & (df_rfm['Quantity'] > 0)]
```

```python
# Verificando os duplicados

df_duplicados = df_rfm[df_rfm.duplicated()]
print(df_duplicados)
```

```python
# Identificar linhas duplicadas
linhas_duplicadas = df_rfm[df_rfm.duplicated()]

# Excluindo linhas duplicadas

df_rfm = df_rfm.drop_duplicates()
```

```python
# Convertendo valores

df_rfm['CustomerID'] = df_rfm['CustomerID'].astype(int)
df_rfm['InvoiceDate'] = pd.to_datetime(df_rfm['InvoiceDate'])

# Conferindo novos valores
df_rfm.info()
```

```python
# Outliers p/ excluir

df_rfm = df_rfm.drop(df_rfm[(df_rfm['Quantity'] > 10000) & (df_rfm['UnitPrice'] > 5000)].index)
```

```python
# Criando Coluna

df_rfm['Total'] = df_rfm['Quantity'] * df_rfm['UnitPrice']

df_rfm.head()
```

```python
# Calculando a última data de compra

ultima_compra = df_rfm['InvoiceDate'].max()

print(ultima_compra)
```

```python
# Grafico top 10 paises com maior vendas

  # Soma por paises

soma_paises = df_rfm.groupby('Country')['Total'].sum()

top_10_paises = soma_paises.sort_values(ascending=False).head(10)


  # Criando Grafico
top_10_paises.plot(kind='bar', x='Coluna_X', y='Coluna_Y', title='Top 10 Paises com Maior Venda')

plt.show()


```

```python
# Grafico top 10 produtos mais vendidos

  # Soma por Produto

soma_produtos = df_rfm.groupby('StockCode')['Total'].sum()

top_10_produtos = soma_produtos.sort_values(ascending=False).head(10)

  # Criando Grafico

top_10_produtos.plot(kind='bar', x='Coluna_X', y='Coluna_Y', title='Top 10 Produtos com Maior Venda', color = 'Green')
plt.show()
```

```python
# Criando uma coluna com mes e ano

df_rfm['Ano_Mes'] = df_rfm['InvoiceDate'].dt.to_period('M')

df_rfm
```

```python
# Valor de Vendas por Mes

  # Soma por Mes

soma_mes = df_rfm.groupby('Ano_Mes')['Total'].sum()

venda_mes = soma_mes.sort_values(ascending=False).head(10)


  # Criando Grafico
venda_mes.plot(kind='bar', x='Coluna_X', y='Coluna_Y', title='Valor de Vendas por Mês')

plt.show()
```

```python
# Valor de venda por mes e por pais

paises_meses = df_rfm.groupby(['Country','Ano_Mes'])['Total'].sum()


top_10_paises_mes = paises_meses.sort_values(ascending=False).head(10)


  # Criando Grafico
top_10_paises_mes.plot(kind='bar', x='Coluna_X', y='Coluna_Y', title='Top 10 Paises com Maior Venda por Mês')

plt.show()

```

```python
# RFM

# Pegar a data mais recente em todo o conjunto de dados

data_atual = df_rfm['InvoiceDate'].max()

# Calcular R para cada cliente
Recencia = df_rfm.groupby('CustomerID')['InvoiceDate'].max()
Recencia = data_atual - Recencia
Recencia = Recencia.dt.days  # Converter para número de dias

# 2. Calcular F
Frequencia = df_rfm.groupby('CustomerID')['InvoiceNo'].nunique()

# 3. Calcular a M
ticket = df_rfm.groupby('CustomerID')['Total'].mean()

# Combinar as três métricas em um novo DataFrame
rfm = pd.DataFrame({
    'Recencia': Recencia,
    'Frequencia': Frequencia,
    'Ticket Médio': ticket
})

rfm
```

```python
rfm.to_csv('rfm_data.csv', index=True, encoding='utf-8-sig', sep=',')
```
